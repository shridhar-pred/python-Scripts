```python
import pandas as pd

# List
Person = [ ['Satyam', 21, 'Patna' , 'India' ],
			['Anurag', 23, 'Delhi' , 'India' ],
			['Shubham', 27, 'Coimbatore' , 'India' ]]

#Create a DataFrame object
df = pd.DataFrame(Person,
				columns = ['Name' , 'Age', 'City' , 'Country'])

# display
display(df)
```

```python
Person = [ ['Amit', 21, 'Patna' , 'India' ],
			['XYZ', 23, 'Delhi' , 'India' ],
			['ABC', 27, 'Coimbatore' , 'India' ]]
```

```python
def modified_data(df, data=None):
  columns= ['Name', 'Age', 'City', 'Country']
  my_dict= dict(zip(columns, data))
  temp_df= pd.DataFrame(my_dict, index=list(range(1)))
  df= df.append(temp_df)
  df=df.reset_index(drop=True)
  return df
```

```python
for i in Person:
  df=modified_data(df, data=i)

df
```
